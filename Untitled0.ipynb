📊 Avaliação de todos os modelos

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

def avaliar_modelo(modelo, X_test, y_test):
    y_pred = modelo.predict(X_test)
    y_prob = modelo.predict_proba(X_test)[:,1]

    print(f"\n📌 Modelo: {modelo.__class__.__name__}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.4f}")

    return roc_curve(y_test, y_prob)

# Avaliação + Curva ROC
plt.figure(figsize=(8,6))

for modelo in [model_lr, model_rf, model_xgb]:
    fpr, tpr, _ = avaliar_modelo(modelo, X_test, y_test)
    plt.plot(fpr, tpr, label=modelo.__class__.__name__)

plt.plot([0,1], [0,1], 'k--')
plt.title('Curvas ROC dos Modelos')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
